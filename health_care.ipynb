```python
import pandas as pd
df = pd.read_csv("healthcare.csv")
```

```python
df.info()
```

```python
df["Age"] = pd.to_numeric(df["Age"], errors="coerce")

df["Visit Date"] = pd.to_datetime(df["Visit Date"], errors="coerce")

df["Gender"] = df["Gender"].astype("category")
df["Condition"] = df["Condition"].astype("category")
df["Medication"] = df["Medication"].astype("category")

df["Patient Name"] = df["Patient Name"].astype("string")
df["Blood Pressure"] = df["Blood Pressure"].astype("string")
df["Email"] = df["Email"].astype("string")
df["Phone Number"] = df["Phone Number"].astype("string")
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df["age"] = df["age"].replace("forty", 40)

df["age"] = pd.to_numeric(df["age"], errors="coerce")

df["age"] = df["age"].fillna(df["age"].median())
```

```python
df.isnull().sum()
```

```python
df["condition"] = df["condition"].fillna(df["condition"].mode()[0])
```

```python
df["blood_pressure"] = df["blood_pressure"].fillna(df["blood_pressure"].mode()[0])
df["cholesterol"] = df["cholesterol"].fillna(df["cholesterol"].median())
df["visit_date"] = df["visit_date"].fillna(
    df["visit_date"].mode()[0]
)
```

```python
df["email"] = df["email"].fillna("Not Provided")
df["phone_number"] = df["phone_number"].fillna(
    "Not Provided"
)
df["phone_number"] = (
    df["phone_number"]
    .replace("", pd.NA)
    .fillna("Not Provided")
)
```

```python
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)
```

```python
df = df.drop_duplicates()
```
